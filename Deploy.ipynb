```python
# Runtime -> change runtime type -> GPU
import tarfile
from google.colab import drive
import os
from shutil import copyfile
import scipy.io
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
```

Carichiamo il modello preaddestrato nella variabile *net*:

```python
copyfile('/content/gdrive/MyDrive/Foundations od Deep Learning/FDL_Project/my_net.h5','my_net.h5') #modello con transfer learning
```

```python
net = None
net = tf.keras.models.load_model('my_net.h5')
```

Importiamo il dataset da keras:

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

Ora è possibile eseguire alcune prove con i dati di test. Scegliendo un indice verrà selezionata un'immagine e il modello cercherà di predire la classe correttamente.

```python
index = 500 # choose the index
example = x_test[index]
plt.imshow(example);
```

```python
# Classe reale vs classe prevista
classification = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
example = np.expand_dims(example, axis=0) # aggiungiamo una dimensione fittizia all'inizio (collegata al batch)
yc_pred = net.predict(example)
y_pred = int(np.argmax(yc_pred, axis=1))

print("The image class is:", classification[y_test[index][0]])
print("The image class prediction is:", classification[[y_pred][0]])
```
